```python
import pandas as pd
from pprint import pprint
```

```python
# Make a reference to the .csv file path
csv_path = "Resources/cities.csv"

# Import as a DataFrame
cities_df = pd.read_csv(csv_path, encoding="utf-8")
cities_df.head(10)
```

```python
#Get rid of "Unnamed" column
cities_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
cities_df.head()
```

```python
# Convert DataFrame to HTML
cities_web = cities_df.to_html(index=False)
```

```python
# Results will be read into website code from HTML file
text_file = open("data_source.html", "w")
text_file.write(cities_web)
text_file.close()
```

```python
# Visualize results for reference
pprint(cities_web)
```
